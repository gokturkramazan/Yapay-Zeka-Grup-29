Veri Kümesi Hakkında
Bu veri kümesi, insan bilişsel performansının çeşitli yaşam tarzı faktörleriyle olan ilişkisini kapsamlı bir biçimde analiz etmeye yönelik olarak hazırlanmıştır. Farklı demografik özelliklere sahip 80.000 gözlemden oluşan bu veri seti; veri bilimi, yapay zekâ ve insan davranışları alanlarında araştırma ve modelleme çalışmaları için nitelikli bir kaynak teşkil etmektedir.

**Veri kümesi, aşağıda belirtilen sütunlar doğrultusunda tablo formatında yapılandırılmıştır:**

- User_ID (Kullanıcı Kimliği): Her bir kullanıcıya özgü benzersiz tanımlayıcı.

- Age (Yaş): Kullanıcının yaşı.

- Gender (Cinsiyet): Kullanıcının cinsiyeti (Erkek, Kadın veya Diğer).

- Sleep_Duration (Uyku Süresi): Kullanıcının gecelik ortalama uyku süresi (saat cinsinden).

- Stress_Level (Stres Seviyesi): Kullanıcının, 1 ile 10 arasında kendi beyanına dayalı stres seviyesi.

- Diet_Type (Beslenme Türü): Kullanıcının beslenme alışkanlığı (Vejetaryen, Vejetaryen Olmayan veya Vegan).

- Daily_Screen_Time (Günlük Ekran Süresi): Kullanıcının ekran karşısında geçirdiği ortalama günlük süre (saat cinsinden).

- Exercise_Frequency (Egzersiz Sıklığı): Kullanıcının fiziksel egzersiz yapma düzeyi (Düşük, Orta veya Yüksek).

- Caffeine_Intake (Kafein Tüketimi): Kullanıcının günlük kafein tüketimi (miligram cinsinden).

- Reaction_Time (Tepki Süresi): Bilişsel testler aracılığıyla ölçülen tepki süresi (milisaniye cinsinden).

- Memory_Test_Score (Hafıza Testi Puanı): Kullanıcının hafıza testinden elde ettiği skor (100 üzerinden).

- Cognitive_Score (Bilişsel Performans Skoru): Yapay zekâ tarafından oluşturulan genel bilişsel performans skoru.

- AI_Predicted_Score (Yapay Zekâ Tahmin Skoru): Makine öğrenimi modeli tarafından tahmin edilen bilişsel performans skoru.

## Veri Görselleştirme

## 1) Korelasyon ısı haritası

```python
plt.figure(figsize=(14,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Spectral', fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.tight_layout()
plt.show()
```

## 2) KDE plot: Uyku Süresi ve Cinsiyete Göre Bilişsel Puan

```python
plt.figure(figsize=(12,6))
sns.kdeplot(data=df, x="Sleep_Duration", hue="Gender", fill=True, common_norm=False, alpha=0.6)
plt.title("Sleep Duration Distribution by Gender", fontsize=14)
plt.tight_layout()
plt.show()
```

## 3) Diet vs Cognitive Score

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Diet_Type', y='Cognitive_Score', data=df, palette='Set3')
plt.title("Cognitive Score by Diet Type")
plt.tight_layout()
plt.show()
```

## 4) Caffeine vs Reaction Time

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Caffeine_Intake', y='Reaction_Time', data=df, hue='Gender', palette='cool')
plt.title("Reaction Time vs Caffeine Intake")
plt.tight_layout()
plt.show()
```

## 5) Age vs Cognitive Score

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Age', y='Cognitive_Score', data=df, color='indigo', scatter_kws={'alpha':0.4})
plt.title("Cognitive Score by Age")
plt.tight_layout()
plt.show()
```

## 6) Exercise vs Cognitive Score

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Exercise_Frequency', y='Cognitive_Score', data=df, palette='Pastel1')
plt.title("Cognitive Score by Exercise Frequency")
plt.tight_layout()
plt.show()
```

## 7) Pairplot of core features

```python
core_features = ['Age', 'Sleep_Duration', 'Stress_Level', 'Cognitive_Score']
sns.pairplot(df[core_features], corner=True, diag_kind='kde', palette='coolwarm')
plt.suptitle("Pairwise Feature Distributions", y=1.02)
plt.show()
```

## 8) Boxen plot for Stress Level vs Cognitive Score

## 9) Distribution of Cognitive Score

## 10) Count plot for Exercise Frequency

## 11) 3D Scatter: Memory vs Reaction vs Cognitive

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Memory_Test_Score'], df['Reaction_Time'], df['Cognitive_Score'], c=df['Cognitive_Score'], cmap='viridis')
ax.set_xlabel('Memory Test Score')
ax.set_ylabel('Reaction Time')
ax.set_zlabel('Cognitive Score')
plt.title("3D Plot: Memory vs Reaction vs Cognitive Score")
plt.show()
```

```python
# ✅ SKILL LABELS
def etiketle_beceri(hafiza_skoru):
    if hafiza_skoru < 60:
        return 'düşük'
    elif 60 <= hafiza_skoru < 85:
        return 'orta'
    else:
        return 'yüksek'

df['skill_level'] = df['Memory_Test_Score'].apply(etiketle_beceri)
df.reset_index(inplace=True)
df.rename(columns={'index': 'user_id'}, inplace=True)
df[['user_id', 'skill_level']].to_csv('skill_labels.csv', index=False)

# ✅ COGNITIVE PROFILE
def etiketle_bilissel_profil(row):
    stres = row['Stress_Level']
    uyku = row['Sleep_Duration']
    if stres >= 3 or uyku < 6:
        return 'Yorgun/Stresli'
    elif stres <= 1 and uyku >= 7:
        return 'Dinlenmiş/Odaklı'
    else:
        return 'Normal'

df['cognitive_profile'] = df.apply(etiketle_bilissel_profil, axis=1)
df[['User_ID', 'cognitive_profile', 'Stress_Level', 'Sleep_Duration']].to_json('cognitive_profile.json', orient='records', indent=4)

print("✅ Dosyalar başarıyla oluşturuldu: skill_labels.csv ve cognitive_profile.json")
```
